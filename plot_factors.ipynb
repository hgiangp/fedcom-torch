```python
from parse_log import * 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
file_name = './logs/server_model.log'
rho = parse_diversity_factor(file_name)
rho_mean = np.ones_like(rho) * rho.mean()

ystick = np.arange(0.05, 0.46, 0.05)
ystick = np.append(ystick, np.round(rho.mean(), 2))
print(ystick)
# ystick = np.concatenate((np.arange(0.05, 0.46, 0.05)[:], np.round(rho.mean(), 2)))
plt.figure()
plt.plot(rho, ':', label='rho')
plt.plot(rho_mean)
plt.grid()
plt.legend()
plt.yticks(ystick)
plt.savefig('rho.png')
plt.show()
```

```python
file_name = './logs/server_model.log'
lipschitz = parse_Lipschitz_factor(file_name)
convex = parse_convex_factor(file_name)
lipschitz_mean = np.ones_like(lipschitz) * lipschitz.mean()
convex_mean = np.ones_like(convex) * convex.mean()

ystick = [2, 3, 4, np.round(lipschitz.mean(), 2), 6, np.round(convex.mean(), 2), 8, 9, 10]
plt.figure()
plt.plot(lipschitz, '--', label='Lipschitz')
plt.plot(convex, ':', label='strongly convex')
plt.plot(lipschitz_mean)
plt.plot(convex_mean)
plt.grid()
plt.legend()
plt.yticks(ystick)
plt.savefig('lischitz2.png')
plt.show()
plt.close()
```
